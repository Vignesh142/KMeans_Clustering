```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/drive/MyDrive/Rinex May Batch/Mall_Customers.csv")
df
```

```python
X=df.iloc[:,3:5].values
```

```python
X
```

```python
from sklearn.cluster import KMeans
k=range(2,14)  #min value is 2, max value is sq root of 200(totalmpoints)=14
SSE=[]  #inertia or sum of squared errors
for i in k:
  km=KMeans(n_clusters=i,random_state=13)
  km.fit(X)
  SSE.append(km.inertia_)
plt.plot(k,SSE)
plt.show()
```

```python
#As per the above graph, elbow is prominent at k=5, hence we will decide on 5 clusters
```

```python
k=5
```

```python
from sklearn.cluster import KMeans
model=KMeans(n_clusters=k,random_state=13)
model.fit(X)
```

```python
y=model.predict(X)
y
```

```python
plt.figure(figsize=(12,5))
plt.scatter(X[y==0,0],X[y==0,1],label="Cluster 0")
plt.scatter(X[y==1,0],X[y==1,1],label="Cluster 1")
plt.scatter(X[y==2,0],X[y==2,1],label="Cluster 2")
plt.scatter(X[y==3,0],X[y==3,1],label="Cluster 3")
plt.scatter(X[y==4,0],X[y==4,1],label="Cluster 4")
plt.xlabel("Annual Income in k$")
plt.ylabel("Spending Score(1-100)")
plt.title("Market Basket Analysis")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,5))
for i in range(k):
  plt.scatter(X[y==i,0],X[y==i,1],label=("Cluster",i))
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],label="Centroids",color="cyan",s=200,alpha=0.5)
plt.xlabel("Annual Income in k$")
plt.ylabel("Spending Score(1-100)")
plt.title("Market Basket Analysis")
plt.legend()
plt.show()
```
